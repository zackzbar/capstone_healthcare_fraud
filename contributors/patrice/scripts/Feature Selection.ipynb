### Read all files

```python
provider = pd.read_csv(os.path.join(file_path,'Train-1542865627584.csv'))
in_patient = pd.read_csv(os.path.join(file_path,'Train_Inpatientdata-1542865627584.csv'))
out_patient = pd.read_csv(os.path.join(file_path,'Train_Outpatientdata-1542865627584.csv'))
benefit = pd.read_csv(os.path.join(file_path,'Train_Beneficiarydata-1542865627584.csv'))
```

### Concat all files(in_patient, out_patient and labeled Provider). 

```python
missing_col_out = list(set(in_patient.columns.to_list()) - set(out_patient.columns.to_list()))

out_patient = out_patient.reindex(columns=out_patient.columns.to_list() + missing_col_out)

all_merge = pd.concat([in_patient,out_patient], axis=0)

```

```python
out_patient = out_patient.merge(provider,how='inner', on=['Provider'])
```

```python
all_merge = all_merge.merge(provider,how='inner', on=['Provider'])
```

```python
all_merge.head()
```

```python
prv_g = all_merge.groupby(['Provider'], as_index=False).agg({'ClaimID':'nunique', 'BeneID':'nunique',
                                                              'InscClaimAmtReimbursed':'mean' }) 
prv_g = prv_g.rename(columns={'BeneID':'MergeTotalPatient','ClaimID':'MergeTotalClaim',
                              'InscClaimAmtReimbursed':'MergeAvgClaimReimbursed' })
prv_g = prv_g.merge(provider,how='inner', on=['Provider'])
```

```python
prv_g.head()
```

```python
top_f = all_merge[all_merge.ClmAdmitDiagnosisCode.isin(['78650', '78605', '486', '4280', '7802', 'V5789', '78097', '78900',
       '78079', '49121', '42731', '71536', '5990', '5789', '78060',
       '0389', '78659', '41401', '78609', '43491'])]
```

```python
topf_g = top_f.groupby(['Provider'], as_index=False).agg({'ClaimID':'nunique'}).reset_index() 
topf_g = topf_g.rename(columns={'ClaimID':'NumTopCodeClaim'}).drop('index', axis=1)
prv_g =prv_g.merge(topf_g, how='left',on='Provider')
```

```python
phy = all_merge.groupby('AttendingPhysician')['Provider'].size().reset_index()

phy = phy[phy['Provider']>1]

dup = all_merge[all_merge.AttendingPhysician.isin(phy.AttendingPhysician)]

dup_g = dup.groupby('Provider')['ClaimID'].size().reset_index()

dup_g['SharePhysician'] = 1


prv_g1 = dup_g.merge(prv_g,how='right',on=['Provider']).drop('ClaimID',axis=1)

in_g = in_patient.groupby(['Provider'], as_index=False).agg({'ClaimID':'count','InscClaimAmtReimbursed':'mean' }) 
in_g = in_g.rename(columns={'ClaimID':'InTotalClaim','InscClaimAmtReimbursed':'InAvgClaimReimbursed'})#.drop(['index'], axis=1)

out_g = out_patient.groupby(['Provider'], as_index=False).agg({'ClaimID':'count','InscClaimAmtReimbursed':'mean' }) 
out_g = out_g.rename(columns={'ClaimID':'OutTotalClaim','InscClaimAmtReimbursed':'OutAvgClaimReimbursed'})#.drop(['index'], axis=1)

prv_g1 = prv_g1.merge(in_g,how='left',on=['Provider'] )
prv_g1 = prv_g1.merge(out_g,how='left',on=['Provider'] )
prv_g1.fillna(0, inplace=True) #= prv_g1.SharePhysician.fillna(0)

prv_g1.head(5)
```
